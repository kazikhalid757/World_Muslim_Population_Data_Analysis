```python
# Step 1: Define the start and end year
start_year = '1960'
end_year = '2021'

# Step 2: Calculate the growth rate for each country
def calculate_growth_rate(row):
    initial_value = row[start_year]
    final_value = row[end_year]
    return ((final_value - initial_value) / initial_value) * 100

# Step 3: Apply the function to create a new column for the growth rate
df['Growth_Rate'] = df.apply(calculate_growth_rate, axis=1)

df.head()
```

```python
#try to display top highest_growth_rates country

grouped_df = df.groupby(['Country Name', 'Growth_Rate'])

highest_growth_rates = grouped_df['Growth_Rate'].max()

sorted_highest_growth_rates = highest_growth_rates.sort_values(ascending=False)[:5]

display(sorted_highest_growth_rates)




Country = ['Qatar', 'United Arab Emirates', 'Sint Maarten (Dutch part)', 'Kuwait', 'Jordan']
Growth_Rate = [7288.31, 6918.99, 1519.28, 1291.59, 1214.75]

# Sort data in ascending order
sorted_indices = sorted(range(len(Growth_Rate)), key=lambda k: Growth_Rate[k])
Country_sorted = [Country[i] for i in sorted_indices]
Growth_Rate_sorted = [Growth_Rate[i] for i in sorted_indices]

# Create a bar chart with adjusted figure size and different colors
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(Country_sorted, Growth_Rate_sorted, color=['b', 'g', 'r', 'c', 'm'], label='Growth Rate')

ax.set_ylabel('Growth Rate')
ax.set_title('Top 5 Countries with Highest Growth Rates Over Time')
ax.legend(title='Countries')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Replace this with your actual dataset
data = pd.read_csv("World_Muslim_Population _Dataset.csv")

# Function to create the line chart for a selected country
def plot_population(country_name):
    country_data = data[data['Country Name'] == country_name].values[0]
    years = data.columns[2:-1]
    population = country_data[2:-1]  # Exclude 'Growth_Rate' from population values
    
    plt.figure(figsize=(10, 6))
    plt.plot(years, population, marker='o')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title(f'Population for {country_name}')
    plt.xticks(rotation=90)

    # Annotate each data point with its population value
    for x, y in zip(years, population):
        plt.annotate(f'{y:,}', (x, y), textcoords="offset points", xytext=(0, 10), ha='center')

    st.pyplot(plt)

# Main Streamlit app
def main():
    st.title('Country Population Visualization')
    
    # Dropdown to select the country
    country_list = data['Country Name'].unique().tolist()
    selected_country = st.selectbox('Select a country', country_list)
    
    # Display the line chart for the selected country
    plot_population(selected_country)

if __name__ == '__main__':
    main()
```
